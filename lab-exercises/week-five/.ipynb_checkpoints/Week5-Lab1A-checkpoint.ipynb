```python
newword= 'oxygen'
counter=0 #when counting, we start at zero so when we add 1 we count in order of 1-
for char in newword:
    counter=counter+1
    print(counter, ":", char)
```

```python
odds = [1 ,3, 5, 7]
print('odds are: ', odds)
```

```python
print('first element:', odds[0]) # getting the first element of any list 
```

```python
print('last element:', odds[len(odds)-1]) # getting the last element of any list 
```

```python
print('last element:', odds[-1]) # getting the last element of any list 
```

```python
for numbers in odds:
    print(numbers)
```
